### On remaque bien que la ligne n'est pas très pérformante pour notre ensemble de données, il faut optimiser d'avantage ce model, nous allons par la suite calculer la fonction coût sans vectorisation de notre model dans un premier temps.

```python
def CoutWithoutVectorization(X, y , teta):
    m = y.shape[0]
    somme  = 0
    for i in range(y.shape[0]):
        somme  = somme + np.sum((model(X, teta) - y)**2 )
    return 1/m * somme
```

```python
cout_no_vectorization = CoutWithoutVectorization(X, y,teta)
```

#### Le coût de notre model est énorme, on va rechercher à le diminuer le mieux que possible

```python
cout_no_vectorization
```

#### Fonction pour la descente de gradient:
##### Dans cette fonction nous auront besoin de :
###### De calculer le gradient de notre fonction coût. 
###### Notre learning_rate (taux d'apprentissage) $\alpha$ : représente notre déplacement pour avoir les meilleurs paramètres 
$\theta{1}$ et $\theta{2}$.

```python
def Gradient(X, y,teta):
    m = y.shape[0]
    return 1/m * X.T.dot(model(X, teta) - y)
```

```python
Gradient(X, y, teta)
```

```python
def GradientDescente(X, y, teta , alpha , iteration):
    cout_x = np.zeros(iteration)
    for i in range(0, iteration):
        teta = teta - (alpha * Gradient(X,y, teta))
        ## Calculer pour chaque modification de theta la fonction coût 
        cout_x[i] = CoutWithoutVectorization(X,y,teta)
    return teta, cout_x
```

```python
nombre_iteration = 1500
new_teta , mes_couts= GradientDescente(X, y, teta, 0.01, nombre_iteration)
```

```python
learning_rate = [0.02, 0.05,0.001,0.005,0.1,0.09,0.009]
```

```python
import warnings
def CompareLearningRate(learning_rate, X,y, teta, iteration):
    global_cost = 0
    save_teta = teta
    i = 0
    for rate in learning_rate:
        ## Calculer la descente de gradient, et les coûts
        actual_teta, actual_cost_vector = GradientDescente(X,y,teta, rate, iteration= iteration)
        ### Calculer le cout global
        global_cost = CoutWithoutVectorization(X,y,actual_teta)
        print("Le coût global pour un taux d'apprentissage = ", rate," est = ", global_cost)
        ### Afficher deux figures, une pour la ligne et une autre pour le coût
        ###intilaiser une figure
        plt.figure(figsize=(10,10))
        plt.scatter(X[:,1], y, marker = "x")
        plt.plot(X[:,1],model(X, actual_teta), color= "green", label="Application of the Gradient Descent with a rate equal to "+str(rate))
        plt.plot(X[:,1], model(X, save_teta),   color = "black",label="Without appliying of the Gradient Descent")
        plt.legend()
        plt.show()
        plt.figure(figsize=(10,10))
        plt.plot(range(iteration), actual_cost_vector)
        plt.title("Courbe de la fonction coût avec un rate de "+str(rate)+" et ayant un coût global = "+str(global_cost) )
        plt.show()
        i = i + 1
```
